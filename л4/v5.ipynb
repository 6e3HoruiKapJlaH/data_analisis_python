```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

path = "StudentsPerformance.csv"
data = pd.read_csv(path)
print(data.head())
```

```python
data["race/ethnicity"].value_counts().plot(kind = "bar").set_title("Number of Students in Each race")
plt.show()
```

```python
data.groupby("race/ethnicity")["math score"].max().plot(kind = "bar").set_title("Max Math")
plt.show()
```

```python
df2 = data.groupby(['race/ethnicity', 'gender'])['writing score'].mean().unstack('gender').fillna(0)
df2[['female', 'male']].plot(kind='bar', stacked = False).set_title("Average Writing Score")
plt.show()
```

```python
plt.hist([data[data['test preparation course']=="completed"]["reading score"], data[data['test preparation course']=="none"]["reading score"]], bins=15)
plt.legend(["Completed", "None"])
plt.show()
```

```python
#plot histogram of scores sum for students who completed the test prep course and those who did not
df = data[data["test preparation course"] == 'completed']
df2 = data[data["test preparation course"] == 'none']
plt.hist([df["math score"] + df["reading score"] + df["writing score"], df2["math score"] + df2["reading score"] + df2["writing score"]], bins=15)
plt.legend(["Completed", "None"])
```

```python
#plot hisogram of reading scores
plt.hist(data["reading score"], bins='auto')
plt.show()
```

```python

lvl_of_education = data["parental level of education"].unique()
plt.hist([data.loc[data["parental level of education"] == x, "math score"] for x in lvl_of_education], label=lvl_of_education, bins=12)
plt.legend(lvl_of_education)
plt.show()

plt.boxplot(data["math score"])
```

```python
""" -1 = perfectly negative linear correlation between the two variables
0 = no linear correlation between the two variables
1 = a perfectly positive linear correlation between the two variables. """
```

```python
plt.scatter( data["reading score"], data["writing score"], 
            c =  data["reading score"],cmap = 'gist_rainbow_r',
            marker = '.')

plt.title('corelation between reading and writing scores')
data["reading score"].corr(data["writing score"])
```

```python
plt.scatter(  data["math score"], data["reading score"],
            c =  data["math score"],cmap = 'gist_rainbow_r',
            marker = '.')

plt.title('corelation between reading and writing scores')
data["math score"].corr(data["reading score"])
```

```python
plt.scatter(  data["math score"], data["writing score"],
            c =  data["math score"],cmap = 'gist_rainbow_r',
            marker = '.')

plt.title('corelation between reading and writing scores')
data["math score"].corr(data["writing score"])
```
